```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Generate sample data (you can replace this with your own data)
np.random.seed(42)
X = np.random.rand(100, 1) * 10  # Feature (independent variable)
y = 2.5 * X + np.random.randn(100, 1) * 2 + 5  # Target (dependent variable) with some noise
```

```python
# Visualize the data
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='green', label='Data points')
plt.title('Single Variable Linear Regression')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (y)')
plt.legend()
plt.show()
```

```python
# Create and train the linear regression model
model = LinearRegression()
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)
```

```python
# Model evaluation
print(f"Coefficient (slope): {model.coef_[0][0]:.4f}")
print(f"Intercept: {model.intercept_[0]:.4f}")
print(f"Mean squared error: {mean_squared_error(y, y_pred):.4f}")
print(f"Coefficient of determination (R²): {r2_score(y, y_pred):.4f}")
```

```python
# Plot the regression line
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, y_pred, color='red', linewidth=2, label='Regression line')
plt.title('Single Variable Linear Regression with Fitted Line')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (y)')
plt.legend()
plt.show()
```
